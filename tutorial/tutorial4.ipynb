# Tutorial 4: Implementation of a reservoir into the Lahn example project

```python
import os
import hydpy
from hydpy import print_vector
hp = hydpy.HydPy("HydPy-H-Lahn_dam")
hydpy.pub.timegrids = "1990-01-01", "2000-12-31", "1d"
hp.prepare_everything()
init_conditions = hp.conditions
hp.simulate()
```
